## 1. MENGAKSES DATABASE

```python
# buat variabel cursor dari koneksi ke database kita
cursor = conn.cursor()
```

```python
# tuliskan query yang akan kita jalankan
# query di bawah akan menampilkan tabel - tabel yang ada di database kita
query = """
SELECT name FROM sqlite_master WHERE type='table' AND name NOT LIKE 'sqlite_%';
"""
```

```python
# menjalankan query dan menampilkan hasil
cursor.execute(query)
print(cursor.fetchall())
```

## 2. WHERE STATEMENT

```python
# memfilter data yang memiliki quantity > 50
query_where = """

SELECT productid as product_id, quantity FROM orderdetails
WHERE quantity > 50

"""
pd.read_sql_query(query_where, conn)
```

## 3. ORDER STATEMENT

```python
# mengurutkan data (Z to A) berdasarkan quantity
query_order = """

SELECT productid as product_id, quantity FROM orderdetails
ORDER BY 2 DESC

"""
pd.read_sql_query(query_order, conn)
```

## 4. LIMIT STATEMENT

```python
# mengambil 10 data dari tabel order details
query_limit = """SELECT * FROM orderdetails
LIMIT 10

"""
pd.read_sql_query(query_limit, conn)
```

## 5. MENGGABUNGKAN STATEMENT (WHERE,ORDER,LIMIT)

```python
# memfilter data yang memiliki quantity > 50
# mengurutkan data (Z to A) berdasarkan quantity
# mengambil 10 data dari tabel order details
query_gab = """

SELECT productid as product_id, quantity FROM orderdetails
WHERE quantity > 50
ORDER BY 2 DESC
LIMIT 10

"""
pd.read_sql_query(query_gab, conn)
```

```python
# apa yang akan terjadi jika penulisan STATEMENT where, order, limit, tidak terurut?
query_gab = """

SELECT productid as product_id, quantity FROM orderdetails
ORDER BY 2 DESC
WHERE quantity > 50
LIMIT 10

"""
pd.read_sql_query(query_gab, conn)
```

### 6.1. Menganalisa aktivitas pemesanan

```python
# tampilkan data pemesanan pada customer id VINET dengan tujuan negara France, dan memiliki Freight kurang dari 3
query_limit = """SELECT * FROM Orders WHERE CustomerID = 'VINET' AND ShipCountry = 'France' AND Freight <3


"""
pd.read_sql_query(query_limit, conn)
```

```python
# tampilkan data pemesanan dengan pengecualian --> ship region tidak boleh bernilai NONE
# dan diurutkan dari nilai freight terendah dengan rentang nilai freight antara 30 sampai 35
query_limit = """SELECT * FROM Orders
                  WHERE ShipRegion IS NOT NULL
                  AND Freight BETWEEN 30 AND 35
                  ORDER BY Freight ASC


"""
pd.read_sql_query(query_limit, conn)
```

```python
# tampilkan data order pada ship address Tsawassen
query_limit1 = """SELECT * FROM Orders WHERE ShipAddress LIKE '%Tsawassen%'


"""
pd.read_sql_query(query_limit1, conn)
```

### 6.2 Mengidentifikasi Customer

```python
# tampilkan data customer yang memiliki jabatan sebagai manager dan tinggal di Perancis dan USA
query_limit2 = """
               SELECT * FROM Customers WHERE ContactTitle LIKE '%Manager%'
               AND Country in ('France', 'USA')


"""
pd.read_sql_query(query_limit2, conn)
```

```python
# tampilkan sample 30 data customer di USA
query_limit3 = """SELECT * FROM Customers WHERE Country = 'USA' LIMIT 30


"""
pd.read_sql_query(query_limit3, conn)
```
